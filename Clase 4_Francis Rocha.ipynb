# Funciones Matemáticas: Imagen y Preimagen | Principios de Demostración

## Introducción a las Funciones Matemáticas

Las funciones matemáticas son una parte fundamental de las matemáticas, necesarias para entender y formalizar diversas relaciones entre conjuntos. Se destaca la importancia de una comprensión sólida de los conceptos básicos para evitar problemas en el estudio de temas más avanzados.

### Concepto General de Función
Una función es una relación entre dos conjuntos que asocia a cada elemento del primer conjunto exactamente un elemento del segundo conjunto. Esta relación no se limita a números; puede involucrar otros tipos de conjuntos, como palabras o letras del alfabeto.

#### Ejemplo:
Considera una función que asocia palabras en español con la primera letra de cada palabra. Esta función no numérica muestra la flexibilidad de las funciones en matemáticas.

## Imagen y Preimagen de una Función

### Imagen
La **imagen** de un conjunto bajo una función es el conjunto de todos los valores que la función toma al aplicar sus elementos.

#### Ejemplo:
Si tenemos una función $ f $ que asocia a cada número natural su doble, y consideramos el conjunto de los números naturales $ \{1, 2, 3, \ldots\} $, la imagen de este conjunto bajo $ f $ sería $ \{2, 4, 6, \ldots\} $.

### Preimagen
La **preimagen** de un conjunto bajo una función es el conjunto de todos los elementos del dominio que son mapeados a elementos del conjunto dado en el codominio.

#### Ejemplo:
Si $ f(x) = x^2 $ y queremos encontrar la preimagen del conjunto $ \{4, 9\} $, necesitamos hallar todos los $ x $ tales que $ f(x) \in \{4, 9\} $. Así, la preimagen sería $ \{-3, -2, 2, 3\} $.

## Propiedades y Relaciones

### Contención de la Imagen
La imagen de un conjunto bajo una función está contenida en el espacio de llegada, pero no necesariamente es igual a él. Esto es conocido como el rango de la función.

#### Ejemplo:
Para una función $ f(x) = e^x $, si consideramos el conjunto de todos los números reales $ \mathbb{R} $ como el dominio, la imagen será el conjunto de todos los números reales positivos $ (0, \infty) $.

### Preimagen y Conjuntos Vacíos
Es importante recordar que la preimagen de ciertos valores puede ser el conjunto vacío si no existe ningún elemento en el dominio que se mapee a esos valores bajo la función dada.

#### Ejemplo:
Si consideramos la función $ f(x) = x^2 $ y el valor -10, la preimagen de -10 es el conjunto vacío porque no hay ningún número real $ x $ tal que $ x^2 = -10 $.

## Aplicaciones y Ejemplos

### Secuencias Numéricas
Una secuencia es un tipo particular de función que mapea números naturales a otros números. Por ejemplo, la secuencia que asocia a cada número natural su doble puede ser representada como una función $ f(n) = 2n $.

### Funciones Trigonométricas
Las funciones trigonométricas también ofrecen ejemplos claros de imagen y preimagen. Consideremos la función seno, $ G(x) = \sin(x) $. La imagen de esta función es el intervalo $ [-1, 1] $, ya que esos son los valores que el seno puede tomar. La preimagen de un valor como 0 sería todos los $ x $ tales que $ \sin(x) = 0 $, es decir, $ x = n\pi $ para $ n $ entero.

## Principios de Demostración en Funciones

### Importancia de la Precisión
Al trabajar con funciones, es crucial utilizar definiciones precisas y recordar propiedades fundamentales. Esto es especialmente importante en demostraciones matemáticas, donde la exactitud y el rigor son esenciales.

#### Ejemplo:
Si queremos demostrar que la preimagen de un valor específico es el conjunto vacío, debemos usar la definición exacta de la función y verificar que no hay ningún elemento en el dominio que se mapee a ese valor.

## Conclusión

Entender las imágenes y preimágenes de funciones es esencial para el estudio de las matemáticas. Estos conceptos no solo ayudan a formalizar relaciones entre conjuntos, sino que también son cruciales en diversas aplicaciones matemáticas y en la resolución de problemas complejos. Mantener una base sólida en estos conceptos y revisarlos regularmente asegura una comprensión profunda y duradera.


```python
# Importamos las bibliotecas necesarias
import numpy as np
import matplotlib.pyplot as plt

# Definimos una función que asocia a cada número su doble
def f(x):
    return 2 * x

# Creamos un conjunto de números naturales
naturales = np.array([1, 2, 3, 4, 5])

# Calculamos la imagen de estos números bajo la función f
imagen = f(naturales)

# Imprimimos el conjunto de números naturales y su imagen
print("Números naturales:", naturales)
print("Imagen bajo f(x) = 2x:", imagen)

# Visualizamos la función y su imagen
plt.figure(figsize=(10, 5))
plt.plot(naturales, imagen, 'ro', label='Imagen de naturales')
plt.xlabel('x')
plt.ylabel('f(x) = 2x')
plt.title('Función f(x) = 2x')
plt.legend()
plt.grid(True)
plt.show()

# Definimos una función que asocia a cada número su cuadrado
def g(x):
    return x**2

# Definimos un conjunto de valores para los que queremos calcular la preimagen
valores = np.array([4, 9])

# Calculamos la preimagen de estos valores bajo la función g
# Buscamos los números que al cuadrado dan 4 y 9
preimagen = [np.sqrt(val) for val in valores] + [-np.sqrt(val) for val in valores]

# Imprimimos los valores y su preimagen
print("Valores:", valores)
print("Preimagen bajo g(x) = x^2:", preimagen)

# Visualizamos la función g(x) = x^2
x = np.linspace(-5, 5, 400)
y = g(x)

plt.figure(figsize=(10, 5))
plt.plot(x, y, label='g(x) = x^2')
plt.scatter(preimagen, g(np.array(preimagen)), color='red', zorder=5, label='Preimagen de [4, 9]')
plt.xlabel('x')
plt.ylabel('g(x) = x^2')
plt.title('Función g(x) = x^2')
plt.legend()
plt.grid(True)
plt.show()

# Definimos una función seno
def h(x):
    return np.sin(x)

# Definimos un rango de valores de x
x = np.linspace(-2*np.pi, 2*np.pi, 400)

# Calculamos la imagen de estos valores bajo la función seno
y = h(x)

# Visualizamos la función seno
plt.figure(figsize=(10, 5))
plt.plot(x, y, label='h(x) = sin(x)')
plt.xlabel('x')
plt.ylabel('h(x) = sin(x)')
plt.title('Función h(x) = sin(x)')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True, which='both')
plt.legend()
plt.show()

# Calculamos la preimagen de 0 bajo la función seno
# Los valores de x tales que sin(x) = 0 son múltiplos de pi
preimagen_cero = [n*np.pi for n in range(-2, 3)]

# Imprimimos la preimagen de 0
print("Preimagen de 0 bajo h(x) = sin(x):", preimagen_cero)

# Mostramos la preimagen de 0 en la gráfica del seno
plt.figure(figsize=(10, 5))
plt.plot(x, y, label='h(x) = sin(x)')
plt.scatter(preimagen_cero, h(np.array(preimagen_cero)), color='red', zorder=5, label='Preimagen de 0')
plt.xlabel('x')
plt.ylabel('h(x) = sin(x)')
plt.title('Función h(x) = sin(x)')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True, which='both')
plt.legend()
plt.show()
```
