## **Funciones: Utilidad, Tipos y Composición**

### 1. Concepto y Representación de Función
Una función es una regla de asociación entre elementos de un espacio inicial (dominio) y un espacio de llegada (codominio). Esto se puede representar matemáticamente como $ f: A \rightarrow B $, donde $ A $ es el dominio y $ B $ es el codominio.

#### *Imagen y Preimagen*
- **Imagen**: Es el conjunto de elementos en el codominio que son mapeados desde el dominio. Formalmente, para un conjunto $ S \subseteq A $, la imagen es $ f(S) = \{ f(x) \,|\, x \in S \} $.
- **Preimagen**: Es el conjunto de elementos en el dominio que se mapean a un elemento específico en el codominio. Para un conjunto $ T \subseteq B $, la preimagen es $ f^{-1}(T) = \{ x \in A \,|\, f(x) \in T \} $.

### 2. Tipos de Funciones
Las funciones se pueden clasificar en diferentes tipos según sus propiedades:

#### *2.1. Inyectividad*
Una función $ f $ es inyectiva si elementos distintos en el dominio tienen imágenes distintas en el codominio. Formalmente, $ f(x_1) = f(x_2) $ implica $ x_1 = x_2 $.

#### *2.2. Sobreyectividad*
Una función $ f $ es sobreyectiva si cada elemento del codominio es la imagen de al menos un elemento del dominio. Es decir, $ \forall y \in B, \exists x \in A $ tal que $ f(x) = y $.

#### *2.3. Biyectividad*
Una función es biyectiva si es inyectiva y sobreyectiva a la vez. Esto implica una correspondencia uno a uno y sobre toda entre el dominio y el codominio.

#### 3. Función de Utilidad
En economía, una función de utilidad mide la satisfacción o preferencia de un consumidor por diferentes conjuntos de bienes.

### Representación Gráfica
La función de utilidad de dos bienes se puede representar gráficamente con una superficie de utilidad en lugar de una curva. Esto proporciona una visualización tridimensional de la utilidad obtenida de diferentes combinaciones de bienes.

#### 4. Composición de Funciones
La composición de funciones es el proceso de aplicar una función a los resultados de otra función. Si tenemos dos funciones $ f: A \rightarrow B $ y $ g: B \rightarrow C $, la composición de $ g $ con $ f $ se denota como $ g \circ f $ y se define como $ (g \circ f)(x) = g(f(x)) $.

### Ejemplo de Composición

Si $ f $ mapea elementos de $ A $ a $ B $ y $ g $ mapea elementos de $ B $ a $ C $, entonces $ g \circ f $ mapea elementos de $ A $ a $ C $.

### 5. Igualdad en Funciones
Se presentan tres nociones de igualdad relacionadas con funciones:

#### *5.1. Igualdad de Elementos*

Se refiere a la comparación directa entre números o valores. Dos elementos son iguales si tienen el mismo valor.

#### *5.2. Igualdad de Conjuntos*
Implica la doble contención: dos conjuntos $ A $ y $ B $ son iguales si $ A \subseteq B $ y $ B \subseteq A $.

#### *5.3. Igualdad entre Funciones*
Dos funciones $ f $ y $ g $ son iguales si $ f(x) = g(x) $ para cada $ x $ en el dominio.

### 6. Importancia en Economía
Los conceptos de función, inyectividad, sobreyectividad, y composición son fundamentales en economía para entender relaciones entre diferentes variables económicas, modelar preferencias y comportamientos de consumidores, y analizar el impacto de cambios en políticas o condiciones de mercado.



```python
# Importamos las bibliotecas necesarias
import matplotlib.pyplot as plt
import numpy as np

# Definimos una función simple f(x) = 2x + 3
def f(x):
    return 2 * x + 3

# Definimos otra función simple g(x) = x^2
def g(x):
    return x**2

# Creamos un rango de valores de x para evaluar las funciones
x_values = np.linspace(-10, 10, 400)

# Calculamos los valores de f(x) y g(x)
f_values = f(x_values)
g_values = g(x_values)

# Composición de funciones: g(f(x))
# Evaluamos primero f(x) y luego aplicamos g a esos valores
gf_values = g(f_values)

# Composición de funciones: f(g(x))
# Evaluamos primero g(x) y luego aplicamos f a esos valores
fg_values = f(g_values)

# Graficamos las funciones y sus composiciones
plt.figure(figsize=(12, 8))

# Graficamos f(x)
plt.subplot(2, 2, 1)
plt.plot(x_values, f_values, label='$f(x) = 2x + 3$')
plt.title('Función $f(x)$')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.legend()

# Graficamos g(x)
plt.subplot(2, 2, 2)
plt.plot(x_values, g_values, label='$g(x) = x^2$', color='orange')
plt.title('Función $g(x)$')
plt.xlabel('$x$')
plt.ylabel('$g(x)$')
plt.legend()

# Graficamos la composición g(f(x))
plt.subplot(2, 2, 3)
plt.plot(x_values, gf_values, label='$g(f(x))$', color='green')
plt.title('Composición $g(f(x))$')
plt.xlabel('$x$')
plt.ylabel('$g(f(x))$')
plt.legend()

# Graficamos la composición f(g(x))
plt.subplot(2, 2, 4)
plt.plot(x_values, fg_values, label='$f(g(x))$', color='red')
plt.title('Composición $f(g(x))$')
plt.xlabel('$x$')
plt.ylabel('$f(g(x))$')
plt.legend()

plt.tight_layout()
plt.show()

# Inyectividad: Verificamos si una función es inyectiva
# f(x) = 2x + 3 es inyectiva si para f(x1) = f(x2), entonces x1 = x2
# Demostración matemática: Si 2x1 + 3 = 2x2 + 3, entonces x1 = x2
# En código, podemos probar con dos valores diferentes y ver si sus imágenes son diferentes

x1 = 1
x2 = 2

# Calculamos las imágenes
f_x1 = f(x1)
f_x2 = f(x2)

print("f({}) = {}".format(x1, f_x1))
print("f({}) = {}".format(x2, f_x2))

if f_x1 != f_x2:
    print("f(x) = 2x + 3 es inyectiva porque f({}) != f({})".format(x1, x2))
else:
    print("f(x) = 2x + 3 no es inyectiva porque f({}) = f({})".format(x1, x2))

# Sobreyectividad: Verificamos si una función es sobreyectiva
# f(x) = 2x + 3 es sobreyectiva si para cada y en el codominio existe un x en el dominio tal que f(x) = y
# Demostración matemática: Para cualquier y, podemos encontrar x = (y - 3) / 2
# En código, podemos probar con un valor de y y encontrar su preimagen

y = 7

# Calculamos la preimagen
x_preimagen = (y - 3) / 2

print("Para y = {}, x = {} tal que f(x) = y".format(y, x_preimagen))

# Si f(x_preimagen) es igual a y, entonces la función es sobreyectiva
if f(x_preimagen) == y:
    print("f(x) = 2x + 3 es sobreyectiva porque para y = {}, existe x = {} tal que f(x) = y".format(y, x_preimagen))
else:
    print("f(x) = 2x + 3 no es sobreyectiva")

# Biyectividad: Verificamos si una función es biyectiva
# Una función es biyectiva si es inyectiva y sobreyectiva
# Ya hemos demostrado que f(x) = 2x + 3 es inyectiva y sobreyectiva
# Por lo tanto, f(x) = 2x + 3 es biyectiva
```
